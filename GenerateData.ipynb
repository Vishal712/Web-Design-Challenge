```python
import pandas as pd
```

In order to get the data ready to export to an html code, we must first get it to a proper format with 10 columns.

```python
data_table = pd.read_csv("Resources/city_data.csv")
data_table = data_table.drop(["Unnamed: 0", "Unnamed: 0.1"], axis = 1)
data_table = data_table.reset_index()
data_table = data_table.rename(columns={"index": "City_ID"})
```
